```python
import pandas as pd
```

```python
def cost(data,m,c):
    total_cost=0
    N=len(data)
    for i in range(N):
        x=data.iloc[i,0]
        y=data.iloc[i,1]
        total_cost+=(1/N)*((y-m*x-c)**2)
    return total_cost
```

```python
def update(data,lr,m,c):
    dm=0
    dc=0
    N= len(data)
    for i in range (N):
        x=data.iloc[i,0]
        y=data.iloc[i,1]
        dm+=(-1/N)*(y-m*x-c)*x
        dc+=(-1/N)*(y-m*x-c)
        print(cost(data,m,c))
    return m-lr*dm,c-lr*dc
```

```python
def gda(data,lr,iter):
    m=0
    c=0
    for i in range (iter):
        m,c=update(data,lr,m,c)
    return m,c
```

```python
def run():
    data=pd.read_csv('data.csv',names=['x','y'])
    lr=0.0001
    iter=100
    m,c=gda(data,lr,iter)
    return m,c
```

```python
print(run())
```
